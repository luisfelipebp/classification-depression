```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data-Science/Classificação/depression/me_cfs_vs_depression_dataset.csv')
```

```python
df.head()
```

# Análise Variáveis

```python
columns_numerics = df.select_dtypes(include='number').columns
columns_objects = df.select_dtypes(include='object').columns
```

```python
for col in columns_numerics:
  plt.figure(figsize=(8, 4))
  sns.histplot(data=df, x=col, kde=True)
  plt.title(f'Distribution of {col.replace("_", " ")}')
  plt.xlabel(col)
  plt.ylabel('Count', fontsize=12)
  plt.show()
```

```python
for col in columns_objects:
  plt.figure(figsize=(8, 4))
  sns.countplot(data=df, x=col)
  plt.title(f'Distribution of {col.replace("_", " ")}')
  plt.xlabel(col)
  plt.ylabel('Count', fontsize=12)
  plt.show()
```

```python
for col in columns_numerics:
  plt.figure(figsize=(8, 4))
  sns.boxplot(
        x='diagnosis',
        y=col,
        data=df,
        palette='Set2',
        showmeans=True,
        meanprops={'marker':'o', 'markerfacecolor':'black', 'markeredgecolor':'black'}
    )
  plt.title(f'Distribuição de {col} por Depression')
  plt.xlabel('Status de Depression')
  plt.ylabel(col)
  plt.grid(axis='y', alpha=0.3)
  plt.show()
```
